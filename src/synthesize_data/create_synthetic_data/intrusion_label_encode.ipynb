```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
```

```python
path1 = r"/Users/doth02/Library/CloudStorage/GoogleDrive-doth02@luther.edu/My Drive/02.research/DA/data/intrusion/"

train1 = pd.read_csv(path1 + "onehot_intrusion_train.csv")
test1 = pd.read_csv(path1 + "onehot_intrusion_test.csv")
sdv = pd.read_csv(path1 + "onehot_intrusion_sdv_100k.csv")
gauss = pd.read_csv(path1 + "onehot_intrusion_sdv_gaussian_100k.csv")
cat = pd.read_csv(path1 + "onehot_intrusion_sdv_categorical_100k.csv")

print(train1.shape, train1.columns)
print(test1.shape, test1.columns)

print(sdv.shape, sdv.columns)
print(gauss.shape, sdv.columns)
print(cat.shape, sdv.columns)
```

```python
train1_y = train1['target']
test1_y = test1['target']
sdv_y = sdv['target']
gauss_y = gauss['target']
cat_y = cat['target']
#checking
train1_label = train1_y.unique()
test1_label = test1_y.unique()
sdv_label = sdv_y.unique()
gauss_label = gauss_y.unique()
cat_label = cat_y.unique()

print(len(train1_label))
print(len(test1_label))
print(len(sdv_label))
print(len(gauss_label))
print(len(cat_label))

print(train1_y.value_counts())
print(test1_y.value_counts())
print(sdv_y.value_counts())
print(gauss_y.value_counts())
print(cat_y.value_counts())
```

```python
print(set(test1_label) - set(train1_label))
print(set(test1_label) - set(sdv_label))
print(set(test1_label) - set(gauss_label))
print(set(test1_label) - set(cat_label))
```

```python
encoder = LabelEncoder()
encoder.fit(train1_y)
train1_y_encoded = encoder.transform(train1_y)
test1_y_encoded = encoder.transform(test1_y)
sdv_y_encoded = encoder.transform(sdv_y)
gauss_y_encoded = encoder.transform(gauss_y)
cat_y_encoded = encoder.transform(cat_y)


# #verify by counts

df1 = pd.DataFrame({'train1_y': train1_y.value_counts().values, 'train1_y_encoded': pd.Series(train1_y_encoded).value_counts()})
df2 = pd.DataFrame({'test1_y': test1_y.value_counts().values, 'test1_y_encoded': pd.Series(test1_y_encoded).value_counts()})                 
df3 = pd.DataFrame({'sdv_y': sdv_y.value_counts().values, 'sdv_y_encoded': pd.Series(sdv_y_encoded).value_counts()})
df4 = pd.DataFrame({'gauss_y': gauss_y.value_counts().values, 'gauss_y_encoded': pd.Series(gauss_y_encoded).value_counts()})
df5 = pd.DataFrame({'cat_y': cat_y.value_counts().values, 'cat_y_encoded': pd.Series(cat_y_encoded).value_counts()})


df = pd.concat([df1, df2, df3, df4, df5], axis= 1, join = 'outer')
df
```

```python
#map back
train1['target'] = train1_y_encoded
test1['target'] = test1_y_encoded
sdv['target'] = sdv_y_encoded
gauss['target'] = gauss_y_encoded
cat['target'] = cat_y_encoded
```

```python
#save
#save 4 dataset
train1.to_csv(path1 + "onehot_intrusion_train.csv", index = False)
test1.to_csv(path1 + "onehot_intrusion_test.csv")
sdv.to_csv(path1 + "onehot_intrusion_sdv_100k.csv", index = False)
gauss.to_csv(path1 + "onehot_intrusion_sdv_gaussian_100k.csv", index = False)
cat.to_csv(path1 + "onehot_intrusion_sdv_categorical_100k.csv", index = False)
#done verify by open the csv files in excel
```
