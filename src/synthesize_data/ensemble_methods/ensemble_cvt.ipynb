```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn 
from sklearn.ensemble import RandomForestClassifier
```

```python
df_ori = pd.read_csv('D:\SummerResearch\data\covertype\onehot_covertype_train.csv')
df_test = pd.read_csv('D:\SummerResearch\data\covertype\onehot_covertype_test.csv')
df_ctgan = pd.read_csv('D:\SummerResearch\data\covertype\onehot_covertype_sdv_rf_100k.csv')
df_ori.head()
```

```python
X_ori = df_ori.iloc[:, :-1]
y_ori = df_ori.iloc[:, -1]

X_test = df_test.iloc[:, :-1]
y_test = df_test.iloc[:, -1]

X_ctgan = df_ctgan.iloc[:, :-1]
y_ctgan = df_ctgan.iloc[:, -1]
```

```python
sklearn.__version__
```

```python
rf = RandomForestClassifier()
rf.fit(X_ori, y_ori)
print('Accuracy on training set: {:.3f}'.format(rf.score(X_ori, y_ori)))
print('Accuracy on test set: {:.3f}'.format(rf.score(X_test, y_test)))
```
