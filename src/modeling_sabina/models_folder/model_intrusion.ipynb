```python
df = pd.read_csv('D:\SummerResearch\data\intrusion\pre_label_encode\onehot_intrusion_train.csv')
# df = df.drop(columns=['Unnamed: 0'])

df
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'target' is the column to be encoded
label_encoder = LabelEncoder()

# Fit and transform on the training data
df['target'] = label_encoder.fit_transform(df['target'])
df.to_csv('D:\SummerResearch\data\intrusion\post_label_encode\onehot_intrusion_train.csv')


# Transform on the test data
# Assuming test_dfs is a list containing multiple test dataframes
test_dfs = ['D:\SummerResearch\data\intrusion\pre_label_encode\onehot_intrusion_sdv_categorical_100k.csv', 
            'D:\SummerResearch\data\intrusion\pre_label_encode\onehot_intrusion_sdv_gaussian_100k.csv',
            'D:\SummerResearch\data\intrusion\pre_label_encode\onehot_intrusion_sdv_100k.csv',
            'D:\SummerResearch\data\intrusion\pre_label_encode\onehot_intrusion_test.csv' 
        ]

test_df_names = ['D:\SummerResearch\data\intrusion\post_label_encode\onehot_intrusion_sdv_categorical_100k.csv', 
            'D:\SummerResearch\data\intrusion\post_label_encode\onehot_intrusion_sdv_gaussian_100k.csv',
            'D:\SummerResearch\data\intrusion\post_label_encode\onehot_intrusion_sdv_100k.csv',
            'D:\SummerResearch\data\intrusion\post_label_encode\onehot_intrusion_test.csv' 
        ]

for i, test_df in enumerate(test_dfs):
    test_dfs[i]['target'] = label_encoder.transform(test_df['target'])
    test_dfs[i].to_csv(test_df_names[i])
```
