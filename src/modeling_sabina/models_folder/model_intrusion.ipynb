```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming 'target' is the column to be encoded
label_encoder = LabelEncoder()

# Load the training data and fit the label encoder
df = pd.read_csv('~/Google Drive/My Drive/02.research/DA/data/intrusion/onehot_intrusion_train.csv')
df['target'] = label_encoder.fit_transform(df['target'])

# Save the encoded training data
df.to_csv('~/Google Drive/My Drive/02.research/DA/data/intrusion/post_label_encode/onehot_intrusion_train.csv')

# List of test CSV paths
test_dfs = [
    '~/Google Drive/My Drive/02.research/DA/data/intrusion/onehot_intrusion_sdv_categorical_100k.csv',
    '~/Google Drive/My Drive/02.research/DA/data/intrusion/onehot_intrusion_sdv_gaussian_100k.csv',
    '~/Google Drive/My Drive/02.research/DA/data/intrusion/onehot_intrusion_sdv_100k.csv',
    '~/Google Drive/My Drive/02.research/DA/data/intrusion/onehot_intrusion_test.csv'
]

test_df_names = [
    '~/Google Drive/My Drive/02.research/DA/data/intrusion/post_label_encode/onehot_intrusion_sdv_categorical_100k.csv',
    '~/Google Drive/My Drive/02.research/DA/data/intrusion/post_label_encode/onehot_intrusion_sdv_gaussian_100k.csv',
    '~/Google Drive/My Drive/02.research/DA/data/intrusion/post_label_encode/onehot_intrusion_sdv_100k.csv',
    '~/Google Drive/My Drive/02.research/DA/data/intrusion/post_label_encode/onehot_intrusion_test.csv'
]

# Iterate over the test dataframes, load them, apply the label encoder, and save them
for i, test_df_path in enumerate(test_dfs):
    # Load the test dataframe
    test_df = pd.read_csv(test_df_path)
    
    # Transform the 'target' column using the previously fitted encoder
    test_df['target'] = label_encoder.transform(test_df['target'])
    
    # Save the transformed test dataframe
    test_df.to_csv(test_df_names[i], index=False)
```
