```python
import os
import pandas as pd
import matplotlib.pyplot as plt

output_folder = os.path.abspath(os.path.join(os.getcwd(), '..', '..', 'output', 'adult', 'acc'))

csv_files = [file for file in os.listdir(output_folder) if file.endswith('.csv')]

fig, axs = plt.subplots(len(csv_files), 1, figsize=(10, 5 * len(csv_files)), sharex=True)

for i, csv_file in enumerate(csv_files):
    csv_path = os.path.join(output_folder, csv_file)
    df = pd.read_csv(csv_path)
    
    train_loss = df['train_loss']
    test_loss = df['test_loss']
    
    axs[i].plot(train_loss, label='Train Loss')
    axs[i].plot(test_loss, label='Test Loss')
    
    axs[i].set_title(f'Plot for {csv_file}')
    axs[i].set_xlabel('Epoch')
    axs[i].set_ylabel('Loss')
    axs[i].legend()
    
plt.tight_layout()
save_path = os.path.join(os.getcwd(), 'loss_plots.pdf')
plt.savefig(save_path)
plt.show()

print(f"Plots saved to: {save_path}")
```

## Merging the Results files into one txt file 

```python
output_folder = os.path.abspath(os.path.join(os.getcwd(), '..', '..', 'output', 'adult', 'acc'))

txt_files = [file for file in os.listdir(output_folder) if file.endswith('.txt')]

combined_text = []

for txt_file in txt_files:
    txt_path = os.path.join(output_folder, txt_file)
    with open(txt_path, 'r') as file:
        file_content = file.read()
        combined_text.append(f"===== {txt_file} =====\n")
        combined_text.append(file_content)
        combined_text.append('\n\n')  

combined_file_path = os.path.join(os.getcwd(), 'combined_text_with_headings.txt')
with open(combined_file_path, 'w') as combined_file:
    combined_file.writelines(combined_text)

print(f"Combined text with headings saved to: {combined_file_path}")
```
